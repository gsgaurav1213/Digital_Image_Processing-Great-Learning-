```python
# import libraries
import cv2
```

```python
# import xml files
# haaracscade_frontalface_default.xml
# haaracscade_smile.xml
```

```python
# load the face xml file
cascade_face = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
# open the camera and capture the face
# Open the default camera
cam = cv2.VideoCapture(0)
```

```python
while True:
    # Capture frame-by-frame
    ret, img = cam.read()   # ret= tectangle
    print(ret)
    
    # Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # convert BGR color into Gray scale image
    
    # Detect faces
    faces = cascade_face.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=2,minSize=(30, 30))
    
    for (x, y, w, h) in faces: # x,y = side of rectangle and h,w= height and weight
        cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
 
    # Display the resulting frame
    cv2.imshow('video', img)
    
    # Exit if 'Esc' key is pressed
    if cv2.waitKey(30) & 0xFF == 27:
        break

# Release the camera and close all OpenCV windows
cam.release()
cv2.destroyAllWindows()
```
